```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.random.multivariate_normal(np.zeros(2),
                                         cov=np.identity(2),
                                         size=(500))
data = np.vstack(
        (data, np.random.multivariate_normal(np.zeros(2)+10,
                                             cov=np.identity(2),
                                             size=(300))))
```
